```python
import pandas as pd
```

```python
df = pd.read_excel("576 exterior colours.xlsx")

df1 = pd.read_excel('128 Interior Colours.xlsx')
```

```python
df1
```

```python
df = pd.concat([df, df1], ignore_index=True)
```

```python
df
```

```python
df["Hex"] = df.apply(lambda row: f"#{row['R']:02X}{row['G']:02X}{row['B']:02X}", axis=1)
```

```python
df.sort_values(by='Colour Number')

```

```python
def hex_to_int(hex_value):
    return int(hex_value.strip('#'), 16)

df['Hex_Int'] = df['Hex'].apply(hex_to_int)

# Sort by the new integer column
df_sorted = df.sort_values(by='Hex_Int', ascending=False)

# Drop the Hex_Int column if you don't need it
df_sorted = df_sorted.drop(columns=['Hex_Int'])

# Show the sorted DataFrame
df_sorted
```

```python
df_sorted.to_csv("576_exterior_colours_with_hex.csv", index=False)
```
